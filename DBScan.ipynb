```python
from sklearn.cluster import DBSCAN
import pandas as pd
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_dados = pd.read_csv('housing.csv')
df_dados = df_dados[df_dados.columns[0:2]]
df_dados = df_dados.to_numpy()
```

```python
normalized_data = StandardScaler().fit(df_dados)
df_dados = normalized_data.transform(df_dados)
```

```python
model = DBSCAN(eps = 0.5, min_samples=12, metric='euclidean').\
    fit(df_dados)
```

```python
outliers_df = df_dados[model.labels_ == -1];
cluster_df = df_dados[model.labels_ != -1];
colors = model.labels_
color_outliers = 'grey'
color_cluster_df = colors[colors != -1]
```

```python
sns.scatterplot(x=cluster_df[:,0], y=cluster_df[:,1], c= color_cluster_df, cmap='rainbow')
sns.scatterplot(x=outliers_df[:,0], y=outliers_df[:,1], c= ['grey'])
```
