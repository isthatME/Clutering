```python
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_dados = pd.read_csv('housing.csv')
df_dados = df_dados[df_dados.columns[0:2]]
df_dados = df_dados.to_numpy()

```

```python
distortions = []
for i in range(2, 11):
    model = KMeans(
        n_clusters=i, init='k-means++',
        n_init=10, max_iter=300,
        tol=1e-04, random_state=42
    )
    model.fit(df_dados)
    distortions.append(model.inertia_)
```

```python
plt.plot(range(2, 11), distortions)
plt.xlabel('Número de clusters')
plt.ylabel('Distorção dos dados')
```

```python
sns.scatterplot(x=df_dados[:,0], y=df_dados[:,1], c= model.labels_, cmap='rainbow')
sns.scatterplot(x=model.cluster_centers_[:, 0], y=model.cluster_centers_[:, 1], c=['black'])
```
